# **ANALYSIS SECTION**

## **Impact of region on salary**

```python
%%sql
Select Country, AVG(ConvertedComp) from survey
group by Country
Order by AVG(ConvertedComp) DESC;
```

## **How does Level of education affect salary**

```python
%%sql
select EdLevel, AVG(ConvertedComp) from survey where CompFreq = "Monthly"
group by EdLevel
order by AVG(ConvertedComp) desc;
```

Years of experience in relation to salary

```python
%%sql
select yearscode, AVG(convertedcomp), WorkWeekHrs from survey where CompFreq = "Monthly"
group by yearscode
order by AVG(convertedcomp) desc;
```

##  **Organization Size in relation to salary**

```python
%%sql
select OrgSize, avg(convertedcomp) from survey where CompFreq = "Monthly"
group by OrgSize
order by avg(convertedcomp) desc;
```

## **Impact of Job title to salary**

```python
%%sql
select DevType, country, AVG(convertedcomp), CompFreq from survey where CompFreq = "Monthly"
group by country, CompFreq
order by AVG(Convertedcomp) asc;
```

## **Impact of gender on Salary** 

```python
%%sql
select Gender, country, AVG(convertedcomp), CompFreq from survey where CompFreq = "Monthly"
group by country, CompFreq
order by AVG(convertedcomp) desc;
```

## Impact of sexuality on salary

```python
%%sql
select sexuality, country, AVG(convertedcomp) from survey where CompFreq = "Monthly"
group by country
order by AVG(convertedcomp) desc;
```

## **Impact of ethnicity on salary**

```python
%%sql
select ethnicity, country, AVG(convertedcomp) from survey where CompFreq = "Monthly"
group by country
order by AVG(convertedcomp) desc;
```

## **Types of skills possesed and how it affects the salary**

```python
%%sql
select country,DatabaseWorkedWith,LanguageWorkedWith,NEWCollabToolsWorkedWith, AVG(convertedcomp) from survey where CompFreq = "Monthly"
group by country
order by AVG(convertedcomp) desc limit 15;
```
